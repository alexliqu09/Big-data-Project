Para el etapa del paso de bronze al silver se tomo los archivos parquet y se aplicaron los filters, se trato de unir los datos en un csv y crear las tablas de dimensiones y hechos

```python
%python
from pyspark.sql.functions import lit

df_bronze = None

for i in range(14, 21):
    season = f"{i}-{i+1}"
    temp_df = spark.read.parquet(f"dbfs:/datalake/bronze/epl_temp_{i}_{i+1}/*.parquet")
    temp_df = temp_df.withColumn('season', lit(season))
    if df_bronze is None:
        df_bronze = temp_df
    else:
        df_bronze = df_bronze.union(temp_df)
```

```python
%python
df_bronze.show(2)
```

```python
%python
from pyspark.sql.functions import col

sergio_df = df_bronze.filter(col("player_name") == "Sergio Agüero") \
                    .select("season", "team_title", "goals", "assists", "shots", "key_passes", "npg", "xG", "xA", "xGBuildup", "time")

sergio_df.show()
```

```python
%python
from pyspark.sql.functions import sum

fact_cols = ['player_name', 'season', 'goals', 'assists', 'shots', 'key_passes', 'yellow_cards', 'red_cards']
dim_cols = ['player_name', 'positions', 'team_title']

fact_table = df_bronze.groupby(fact_cols).agg(sum('goals').alias('total_goals'),
                                               sum('assists').alias('total_assists'),
                                               sum('shots').alias('total_shots'),
                                               sum('key_passes').alias('total_key_passes'),
                                               sum('yellow_cards').alias('total_yellow_cards'),
                                               sum('red_cards').alias('total_red_cards'))

dim_table = df_bronze.select(dim_cols).dropDuplicates()

fact_table.join(dim_table, 'player_name', 'inner') \
           .write.format('parquet') \
           .option('path', 'dbfs:/datalake/silver/epl_player_stats/') \
           .save(mode='overwrite')
```

En la capa Gold, se realozó un procesamiento adicional de los datos de la capa Silver para obtener información más significativa y lista para su consumo en informes y análisis.

Primero, se hizó una agregación de los datos de la capa Silver, agrupándolos por jugador y temporada, y calculando diversas estadísticas como la suma total de goles, asistencias, tiros, pases clave, tarjetas amarillas y tarjetas rojas, así como el número de partidos jugados por temporada. Luego, renombré las columnas y calculé estadísticas adicionales por temporada como el promedio de goles por partido, promedio de asistencias por partido, promedio de tiros por partido, promedio de pases clave por partido, promedio de tarjetas amarillas por partido y promedio de tarjetas rojas por partido.

```python
%python
from pyspark.sql.functions import desc

fact_table = spark.read.parquet('dbfs:/datalake/silver/epl_player_stats/*.parquet')

player_stats = fact_table.groupby(['player_name', 'season']).agg({'total_goals': 'sum',
                                                                   'total_assists': 'sum',
                                                                   'total_shots': 'sum',
                                                                   'total_key_passes': 'sum',
                                                                   'total_yellow_cards': 'sum',
                                                                   'total_red_cards': 'sum',
                                                                   'player_name': 'count'})

player_stats = player_stats.sort(['player_name', 'season'])

player_stats = player_stats.withColumnRenamed('sum(total_goals)', 'total_goals') \
                           .withColumnRenamed('sum(total_assists)', 'total_assists') \
                           .withColumnRenamed('sum(total_shots)', 'total_shots') \
                           .withColumnRenamed('sum(total_key_passes)', 'total_key_passes') \
                           .withColumnRenamed('sum(total_yellow_cards)', 'total_yellow_cards') \
                           .withColumnRenamed('sum(total_red_cards)', 'total_red_cards') \
                           .withColumnRenamed('count(player_name)', 'games_played_season') \

player_stats = player_stats.withColumn('goals_per_season', player_stats['total_goals'] / player_stats['games_played_season']) \
                           .withColumn('assists_per_season', player_stats['total_assists'] / player_stats['games_played_season']) \
                           .withColumn('shots_per_season', player_stats['total_shots'] / player_stats['games_played_season']) \
                           .withColumn('key_passes_per_season', player_stats['total_key_passes'] / player_stats['games_played_season']) \
                           .withColumn('yellow_cards_per_season', player_stats['total_yellow_cards'] / player_stats['games_played_season']) \
                           .withColumn('red_cards_per_season', player_stats['total_red_cards'] / player_stats['games_played_season'])

player_stats = player_stats.sort(['season', 'player_name'])

player_stats.write.format('parquet') \
            .option('path', 'dbfs:/datalake/gold/epl_player_stats/') \
            .save(mode='overwrite')
```

```python
%python
from pyspark.sql.functions import col

top_scorers = player_stats.orderBy(col('total_goals').desc())
top_scorers = top_scorers.limit(4)
top_scorers.show()
```
